## Now, Choosing the K- Value:-

```python
error_rate = []

for i in range(1,21):
    
    Knn = KNeighborsClassifier(n_neighbors=i)
    Knn.fit(x_train,y_train)
    pred_i = Knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

## Plotting the graph:-

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,21),error_rate,color='crimson',linestyle='dotted',marker='p',
        markerfacecolor='black',markersize=6)
plt.title('Error Rate vs.K-Value', color = 'green')
plt.xlabel('K',color = 'fuchsia')
plt.ylabel('Error Rate', color = 'darkcyan')
plt.show()
```
