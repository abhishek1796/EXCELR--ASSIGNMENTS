## Importing necessary libraries:-

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import normalize, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier

import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
glass_data = pd.read_csv('glass.csv')
glass_data
```

### Converting the Data to Array for Plotting:-

```python
X = np.array(glass_data.iloc[:,3:5])
y = np.array(glass_data['Type'])
print("Shape of X:"+str(X.shape))
print("Shape of y:"+str(y.shape))
```

## Customize the Color maps:-

```python
cm_dark = ListedColormap(['blue', 'green','red','purple','orange','brown','pink'])
cm_bright = ListedColormap(['navy', 'darkblue','springgreen','peru','bisque','palegreen','teal'])
```

### Creating color Scatter maps:-

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap=cm_dark,s=10,label=y)
plt.show()
```

### Using Seaborn Plots:-

```python
sns.swarmplot(x='Na',y='RI',data=glass_data,hue='Type')
plt.show()
```

### Creating Train Set, Test Set and then Checking the Shapes:-

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,y)
print("Shape of X_Train:"+str(X_train.shape))
print("Shape of y_Train:"+str(Y_train.shape))
print("Shape of X_Test:"+str(X_test.shape))
print("Shape of y_Test:"+str(Y_test.shape))
```

## Now, Using KNN to Classify the glass:-

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,Y_train)
```

## Predicting Results Using KNN fit:-

```python
pred = knn.predict(X_train)
pred
```

## Checking the Accuracy :-

```python
accuracy = knn.score(X_train,Y_train)
print("The accuracy is about :"+str(accuracy))
```

## Computing a Confusion Metrix:-

```python
cnf_matrix = confusion_matrix(Y_train,pred)
print(cnf_matrix)
```

### Plotting the metrix in Pyplot:-

```python
plt.imshow(cnf_matrix,cmap=plt.cm.jet)
plt.show()
```

## Now, Visualization of Confusion Metrix:-

```python
df_cm = pd.DataFrame(cnf_matrix, range(6),range(6))
sns.set(font_scale=1.4)
sns.heatmap(df_cm, annot=True,annot_kws={"size": 16})
plt.show()
```

## Plotting the Decision Boundaries for the Data:-

```python
h = .02  # step size in the mesh
n_neighbors = 5 # No of neighbours
for weights in ['uniform', 'distance']:
    
    # we create an instance of Neighbours Classifier and fit the data.
    clf = KNeighborsClassifier(n_neighbors, weights=weights)
    clf.fit(X, y)

    
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])   # ravel to flatten the into 1D and c_ to concatenate
    
    
    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cm_bright)

    
    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cm_dark,
                edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("3-Class classification (k = %i, weights = '%s')"% (n_neighbors, weights))

plt.show()
```
